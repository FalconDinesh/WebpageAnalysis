```
import sklearn
from sklearn.model_selection import train_test_split
df1=df1.sample(frac=1).reset_index().drop('index',axis=1)
training_dataset, test_dataset = sklearn.model_selection.train_test_split(df1, train_size=0.80, test_size=0.20)
```

```
y_train = training_dataset['rank']
X_train = training_dataset.drop('rank',axis = 1)

y_test = test_dataset['rank']
X_test = test_dataset.drop('rank', axis=1)
```

# SVM

```
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```
